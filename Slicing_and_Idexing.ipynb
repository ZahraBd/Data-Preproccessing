Indexes in pandas are powerful data structures that provide metadata about the data in a DataFrame or Series.



 They act as labels for rows and columns, making it easier to access and manipulate data. Indexes can be simple integer-based indexes or more complex hierarchical indexes.

Slicing allows you to select a subset of rows or columns based on certain criteria, such as label-based indexing, integer-based indexing, or boolean indexing.

Setting and removing indexes

```python
# Index temperatures by city, Setting the index
temperatures_ind = temperatures.set_index("city")

# Look at temperatures_ind
print(temperatures_ind.head(),"\n\n##############")

"""
Reset the index, keeping its contents, Resetting the Index while Keeping it
This line resets the index of the DataFrame temperatures_ind to the default integer index,while keeping its contents.
It prints the resulting DataFrame with the original index as a column and the integer index as the new index.
"""
print(temperatures_ind.reset_index(), "\n\n##############")

"""
Reset the index, dropping its contents
This line resets the index of the DataFrame temperatures_ind to the default integer index and drops its contents,
effectively removing the "city" column from the DataFrame.
It prints the resulting DataFrame with only the integer index.
"""
print(temperatures_ind.reset_index(drop=True),"\n\n##############")
```

Subsetting with .loc[]

The .loc[] accessor in pandas is used for label-based indexing, allowing you to subset rows and columns based on their labels (index names) in a DataFrame. It provides a powerful way to select data from a DataFrame using row and column labels.

```python
# Make a list of cities to subset on
cities = ["Moscow", "Saint Petersburg"]

# Subset temperatures using square brackets
"""
The isin() function in pandas is used to filter data by selecting rows
with specific values in a particular column.
It returns a boolean Series indicating whether each element of the column
is contained in the specified list or array of values.
"""
print(temperatures[temperatures["city"].isin(cities)])

# Subset temperatures_ind using .loc[]
print(temperatures_ind.loc[cities])
```

Setting multi-level indexes



```python
# Index temperatures by country & city
temperatures_ind = temperatures.set_index(["country", "city"])

# List of tuples: Brazil, Rio De Janeiro & Pakistan, Lahore
rows_to_keep = [("Brazil", "Rio De Janeiro"), ("Pakistan", "Lahore")]

# Subset for rows to keep
print(temperatures_ind.loc[rows_to_keep])
```

Sorting by index values

```python
# Sort temperatures_ind by index values
print(temperatures_ind.sort_index())

"""
This line sorts the DataFrame temperatures_ind
by index values at the city level first and then at the country level.
It uses the .sort_index() method with the level parameter set to
a list of index levels to sort by. By default, sorting is in ascending order.
"""
# Sort temperatures_ind by index values at the city level
print(temperatures_ind.sort_index(level=["city","country"]))

# Sort temperatures_ind by country then descending city
print(temperatures_ind.sort_index(level=["country", "city"], ascending=[True, False]))
```

Slicing index values

```python
# Sort the index of temperatures_ind
temperatures_srt = temperatures_ind.sort_index()

# Subset rows from Pakistan to Russia
print(temperatures_srt.loc["Pakistan":"Russia"])

# Try to subset rows from Lahore to Moscow
print(temperatures_srt.loc["Lahore":"Moscow"])

# Subset rows from Pakistan, Lahore to Russia, Moscow
print(temperatures_srt.loc[("Pakistan","Lahore"):("Russia","Moscow")])
```

Slicing in both directions

```python
# Subset rows from India, Hyderabad to Iraq, Baghdad
print(temperatures_srt.loc[("India","Hyderabad"):("Iraq","Baghdad")])

# Subset columns from date to avg_temp_c
print(temperatures_srt.loc[:,"date":"avg_temp_c"])

# Subset in both directions at once
print(temperatures_srt.loc[("India","Hyderabad"):("Iraq","Baghdad"),"date":"avg_temp_c"])
```
